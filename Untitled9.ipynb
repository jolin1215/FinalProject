```python
import pandas as pd

# 讀取Excel檔案
data = pd.read_excel('活頁簿1.xlsx')

# 定義函式，根據讀書時間分配到對應的區間
def assign_time_range(time):
    if time < 5:
        return '3hr'
    elif time == 5:
        return '5hr'
    elif time == 10:
        return '10hr'
    else:
        return '15~hr'

# 新增讀書時間區間欄位
data['讀書時間區間'] = data['每週讀書時間'].apply(assign_time_range)
```

```python
# 指定讀書時間區間的類別順序
time_order = ['3hr', '5hr', '10hr', '15~hr']
data['讀書時間區間'] = pd.Categorical(data['讀書時間區間'], categories=time_order, ordered=True)

# 按照學院和讀書時間區間進行分組，計算人數
grouped_data = data.groupby(['學院', '讀書時間區間']).size().reset_index(name='人數')
```

```python
import matplotlib.pyplot as plt

# 設定圖表樣式
plt.style.use('seaborn')

# 將資料轉換為適合繪製堆疊條形圖的格式
pivot_data = grouped_data.pivot(index='學院', columns='讀書時間區間', values='人數')

# 繪製堆疊條形圖
pivot_data.plot(kind='bar', stacked=True)

# 添加標籤和標題
plt.xlabel('school')
plt.ylabel('number of people')

plt.xticks(rotation=0, ha='right')

# 顯示圖表
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取Excel檔案
data = pd.read_excel('活頁簿1.xlsx')

# 定義函式，根據讀書時間分配到對應的區間
def assign_time_range(time):
    if time <5:
        return '3hr'
    elif time == 5:
        return '5hr'
    elif time == 10:
        return '10hr'
    else:
        return '15~hr'

# 新增讀書時間區間欄位
data['讀書時間區間'] = data['每週讀書時間'].apply(assign_time_range)

# 以時間區間和性別分組，計算每個時間區間的人數
grouped_data = data.groupby(['讀書時間區間', '生理性別']).size().unstack().fillna(0)

# 定義自訂的 x 軸順序
custom_order = ['3hr', '5hr', '10hr', '15~hr']

# 繪製雙曲線圖
grouped_data.plot(kind='line')

# 設定 x 軸刻度的順序
plt.xticks(range(len(custom_order)), custom_order)

# 設定圖表標籤等
plt.xlabel('time of range')
plt.ylabel('number of people')

# 顯示圖表
plt.show()
```
